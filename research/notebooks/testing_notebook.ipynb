```python
import yfinance as yf

tickers = ["^CNXIT", "^CNXPHARMA", "^CNXFMCG", "^NSEBANK", 
           "^CNXAUTO", "^CNXMETAL", "^CNXREALTY", "^CNXENERGY",
           "XLK", "XLV", "XLE", "XLF", "XLY", "XLP", "XLI", "XLB"]

for t in tickers:
    df = yf.download(t, period="max", progress=False)
    if not df.empty:
        print(f"{t:15s} | {df.index[0].date()} → {df.index[-1].date()} | {len(df)} rows")
    else:
        print(f"{t:15s} | NO DATA")
```
